```python
#['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']
list_1 = ["x","y","z"]
comprehension_1 = [[i*j for i in list_1] for j in range(1,5)]
#to flatten this, another list_comprehension is created
comprehension_2 = [val for sublist in comprehension_1 for val in sublist]
print(comprehension_2)
```
